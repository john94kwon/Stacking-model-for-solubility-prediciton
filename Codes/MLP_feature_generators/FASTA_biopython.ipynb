```python
from sklearn import preprocessing
import pandas as pd


#arrange fasta file as sample provided
data = pd.read_csv("C:\\Users\\johnkwon\\Desktop\\[Ultima]_Solubility\\Data\\Generative\\FASTA.csv")
```

```python
X = data.iloc[:,0]  #independent columns
indices=len(X)
```

```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis


list_arom_1=[]
list_insta_1=[]
list_hydropho_1=[]
list_TFR_1=[]
list_AI_1=[]
list_ACPR_1=[]
list_hydrophil_1=[]

for i in range (0,len(X)):
    Y = ProteinAnalysis(X[i])
    A=(Y.count_amino_acids()['A'])
    R=(Y.count_amino_acids()['R'])
    N=(Y.count_amino_acids()['N'])
    D=(Y.count_amino_acids()['D'])
    C=(Y.count_amino_acids()['C'])
    Q=(Y.count_amino_acids()['Q'])
    E=(Y.count_amino_acids()['E'])
    G=(Y.count_amino_acids()['G'])
    H=(Y.count_amino_acids()['H'])
    I=(Y.count_amino_acids()['I'])
    L=(Y.count_amino_acids()['L'])
    K=(Y.count_amino_acids()['K'])
    M=(Y.count_amino_acids()['M'])
    F=(Y.count_amino_acids()['F'])
    P=(Y.count_amino_acids()['P'])
    S=(Y.count_amino_acids()['S'])
    T=(Y.count_amino_acids()['T'])
    W=(Y.count_amino_acids()['W'])
    y=(Y.count_amino_acids()['Y'])
    V=(Y.count_amino_acids()['V'])
    n=len(X[i])

    
    
    from Bio.SeqUtils.ProtParam import ProteinAnalysis
    Aromaticity= Y.aromaticity()
    Instability_index= Y.instability_index()
    Hydrophobic_index= Y.gravy()
    Turn_forming_residue_fraction= ((N+G+P+S)/n)
    Alipathic_index= (A+2.9*V+3.9*I+3.9*L)
    Absolute_charge_per_residue= abs((R+K-D-E)/n-0.03)
    Hydrophilic_index= (3*R+3*D+3*E+3*K+0.3*S+0.2*N+0.2*Q-0.4*T-0.5*A-0.5*H-C-1.3*M-1.5*V-1.8*I-1.8*L-2.3*y-2.5*F-3.5*W)
    
    list_arom_1.append(Aromaticity)
    list_insta_1.append(Instability_index)
    list_hydropho_1.append(Hydrophobic_index)
    list_TFR_1.append(Turn_forming_residue_fraction)
    list_AI_1.append(Alipathic_index)
    list_ACPR_1.append(Absolute_charge_per_residue)
    list_hydrophil_1.append(Hydrophilic_index)
    

    
    
```

```python
df = pd.DataFrame((list_arom_1,list_insta_1,list_hydropho_1,list_TFR_1, list_AI_1, list_ACPR_1,list_hydrophil_1),index=["Aromaticity","Instability index","Hydrophobic index","Turn forming residue fraction","Alipathic index","Absolute charge per residue","Hydrophilic index"])
df1=df.T
```
